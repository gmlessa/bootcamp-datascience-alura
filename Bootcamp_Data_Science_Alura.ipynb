####Desafio 02: adicionar uma coluna mostrando a região de cada estado

```python
##Segunda forma, adicionando uma coluna região ao banco de dados e plotar por região
%%timeit
usaveis["Região"] = 0

for i in usaveis.index:
  if (i[0] == "1"):
    usaveis["Região"][i] = "Norte"
  if (i[0] == "2"):
    usaveis["Região"][i] = "Nordeste"
  if (i[0] == "3"):
    usaveis["Região"][i] = "Sudeste"
  if (i[0] == "4"):
    usaveis["Região"][i] = "Sul"
  if (i[0] == "5"):
    usaveis["Região"][i] = "Centro-Oeste"
```

```python
##Uma alternativa da segunda forma, só que mais performático.
%%timeit
usaveis['Região'] = usaveis.index.map(lambda x: "Norte" if "10"<x<"20" else
                                                           ("Sul" if "40"<x<"50" else
                                                            ("Nordeste" if "20"<x<"30" else
                                                             ("Sudeste" if "30"<x<"40" else "Centro-Oeste"))))
```

```python
#Foram feitas alterações manuais por não saber uma forma melhor de fazer.
dic_estado_populacao = {'35 São Paulo': 46649132,
  '31 Minas Gerais': 21411923,
  '33 Rio de Janeiro': 17463349,
  '29 Bahia': 14985284,
  '41 Paraná': 11597484,
  '43 Rio Grande do Sul': 11466630,
  '26 Pernambuco': 9674793,
  '23 Ceará': 9240580,
  '15 Pará': 8777124,
  '42 Santa Catarina': 7338473,
  '52 Goiás': 7206589,
  '21 Maranhão': 7153262,
  '13 Amazonas': 4269995,
  '32 Espírito Santo': 4108508,
  '25 Paraíba': 4059905,
  '24 Rio Grande do Norte': 3560903,
  '51 Mato Grosso': 3567234,
  '27 Alagoas': 3365351,
  '22 Piauí': 3289290,
  '53 Distrito Federal': 3094325,
  '50 Mato Grosso do Sul': 2839188,
  '28 Sergipe': 2338474,
  '11 Rondônia': 1815278,
  '17 Tocantins': 1607363,
  '12 Acre': 906876,
  '16 Amapá': 877613,
  '14 Roraima': 652713}
```

```python
#Definindo gastos por habitante do estado e sua população
gasto_por_hab_por_estado = [[usaveis["Total"].loc[estado] / dic_estado_populacao[estado], dic_estado_populacao[estado]] for estado in dic_estado_populacao.keys()]
gasto_por_hab_por_estado
```

```python
#Fazendo banco de dados para plotar o gráfico
df_gasto_por_hab = pd.DataFrame(data=gasto_por_hab_por_estado, index=dic_estado_populacao.keys(), columns=['Gasto por habitante', 'Num de Habitantes'])
df_gasto_por_hab
```

```python
#Plotando o gráfico
df_gasto_por_hab.sort_values(by='Num de Habitantes', ascending=False)['Gasto por habitante'].plot(kind='bar', grid=True, figsize=(15,5))
plt.title("Gastos por habitantes dos estados referente a internações no SUS de 2008 a 2020 ordenado por população")
plt.show()
```

```python
#Função para fazer o cálculo por habitante comparado entre dois estados
def custo_hab(populacao_A, estado_A, populacao_B, estado_B):
  gasto_A = usaveis["Total"].loc[estado_A] / populacao_A
  gasto_B = usaveis["Total"].loc[estado_B] / populacao_B
  df_gasto_por_hab = pd.DataFrame(data=[gasto_A, gasto_B], index=[estado_A, estado_B], columns=['Gasto por habitante'])
  ax = df_gasto_por_hab.plot(kind='bar', grid=True, figsize=(15,5))
  ax.set_xticklabels(ax.get_xticklabels(), rotation=0, ha = "right")
  primeiro_estado = estado_A[2:]
  segundo_estado = estado_B[2:]
  plt.title(f"Gasto hospitalar em internações do SUS de 2008 até 2022 por habitante de {primeiro_estado} e {segundo_estado}")
```

```python
#Comparando Goiás e Distrito Federal
custo_hab(df_gasto_por_hab['Num de Habitantes']["52 Goiás"], "52 Goiás", df_gasto_por_hab['Num de Habitantes']["53 Distrito Federal"], "53 Distrito Federal")
```
