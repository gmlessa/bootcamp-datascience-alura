####Desafio 2: faça a mesma análise para o mês mais recente

####Desafio 02: adicionar uma coluna mostrando a região de cada estado

```python
##Segunda forma, adicionando uma coluna região ao banco de dados e plotar por região
%%timeit
usaveis["Região"] = 0

for i in usaveis.index:
  if (i[0] == "1"):
    usaveis["Região"][i] = "Norte"
  if (i[0] == "2"):
    usaveis["Região"][i] = "Nordeste"
  if (i[0] == "3"):
    usaveis["Região"][i] = "Sudeste"
  if (i[0] == "4"):
    usaveis["Região"][i] = "Sul"
  if (i[0] == "5"):
    usaveis["Região"][i] = "Centro-Oeste"
```

```python
##Uma alternativa da segunda forma, só que mais performático.
%%timeit
usaveis['Região'] = usaveis.index.map(lambda x: "Norte" if "10"<x<"20" else
                                                           ("Sul" if "40"<x<"50" else
                                                            ("Nordeste" if "20"<x<"30" else
                                                             ("Sudeste" if "30"<x<"40" else "Centro-Oeste"))))
```

###Ordenando banco de dados

```python
ordenado_por_total = usaveis.sort_values("Total", ascending=False)
ordenado_por_total.drop("Total", axis=1, inplace=True)
ordenado_por_total.head()
```

```python
(ordenado_por_total.head(5).drop("Região", axis=1)).T.plot(figsize=(10,6))
```

####Desafio 03: Formatar o gráfico de custos por mês dos 5 estados, deixando ele agradável (Bonitão, segundo o Gui)

```python
ax = (ordenado_por_total.head(5).drop("Região", axis=1)).T.plot(figsize=(12,6))
ax.set_ylabel("Valores em R$")
ax.set_xlabel("Meses")
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha = "right")
plt.title("Top 5 Unidades da Federação que mais gastaram com procedimentos hospitalares do SUS desde 2008", fontsize=16)
plt.show()
```
